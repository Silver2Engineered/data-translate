```python
import pandas as pd
df = pd.read_csv('./covid-data.csv')
```

```python
df['submission_date'] = pd.to_datetime(df['submission_date'])
df = df.sort_values(['state', 'submission_date'])
df
```

```python
state = 'CA'

df_ca = df[df['state'] == state]

df_ca['tot_cases_diff'] = df_ca['tot_cases'].diff()
df_ca
```

```python
df_ca['tot_cases_diff'].mean()
```

```python
df_ca['new_case_diff'] = df_ca.iloc['new_case'].diff()
print(df_ca['new_case_diff'].mean())
df_ca
```

```python
mean_diff = df_ca['new_case_diff'].mean()

sentence = 'The average new weekly cases in {0} was {1} from {2} to {3}.'.format(
    state, 
    round(mean_diff, 2), 
    df_ca['submission_date'].iloc[0].date(), 
    df_ca['submission_date'].iloc[-1].date())

print(sentence)
```
